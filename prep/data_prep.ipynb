```python
halflives_txt_path = './data/Tani/tani_halflives.txt'
nanoplen = './experiments/hsa_dRNA_HeLa_DRB_0h_1/nanoplen_length_synth_rate.tab'
polyA = './experiments/hsa_dRNA_HeLa_DRB_0h_1/polyA_synth_rate.tab'
save_path = './experiments/hsa_dRNA_HeLa_DRB_0h_1'
```

```python
import pandas as pd

df_nanoplen = pd.read_csv(nanoplen,sep='\t',header=(0))
print(len(df_nanoplen))
df_nanoplen = df_nanoplen[df_nanoplen['conv'] != 0]
print(len(df_nanoplen))
df_nanoplen
```

```python
df_polyA = pd.read_csv(polyA,sep='\t',header=(0))
print(len(df_polyA))

df_polyA = df_polyA[df_polyA['conv'] != 0]
print(len(df_polyA))

df_polyA
```

```python
tani_df = pd.read_csv(halflives_txt_path, sep='\t', header=(0), index_col='Transcript.stable.ID')#, skiprows=(0,1), header=(2))
tani_df[:10]
```

```python
df = df_nanoplen.merge(df_polyA, how="inner", left_on="transcript", right_on="contig", suffixes=("_nanoplen", "_polyA"))
df
```

```python
df = df.merge(tani_df, how="inner", left_on="transcript", right_on="Transcript.stable.ID", suffixes=("", "_tani"))
df
```

```python
df.columns
```

```python
cols = ["transcript", "name", "Gene.stable.ID", "n_nanoplen", "mean_nanoplen", "prop_nanoplen", "min_nanoplen", 'n_polyA', 'mean_polyA', 'sigma_polyA', 'prop_polyA',
       'min_polyA', "t5"]
df = df[cols]
df
```

```python
fpath = save_path+"/features_v1.csv"
df.to_csv(fpath)
```

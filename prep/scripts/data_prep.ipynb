```python
# ### experiment hsa_dRNA_HeLa_DRB_0h_1 ### 
halflives_txt_path = './data/Tani/tani_halflives.txt'
nanoplen = './experiments/hsa_dRNA_HeLa_DRB_0h_1/nanoplen_length_synth_rate.tab'
polyA = './experiments/hsa_dRNA_HeLa_DRB_0h_1/polyA_synth_rate.tab'
save_path = './experiments/hsa_dRNA_HeLa_DRB_0h_1/features_v1.csv'
```

## hsa_dRNA_HeLa_DRB_0h_1

```python
import pandas as pd

df_readlength = pd.read_csv(nanoplen,sep='\t',header=(0))
print(len(df_readlength))
df_readlength = df_readlength[df_readlength['conv'] != 0]
print(len(df_readlength))
df_readlength
```

```python
df_polyA = pd.read_csv(polyA,sep='\t',header=(0))
print(len(df_polyA))

df_polyA = df_polyA[df_polyA['conv'] != 0]
print(len(df_polyA))

df_polyA
```

```python
hl_df = pd.read_csv(halflives_txt_path, sep='\t', header=(0), index_col='Transcript.stable.ID')#, skiprows=(0,1), header=(2))
hl_df[:10]
```

```python
df = df_readlength.merge(df_polyA, how="inner", left_on="transcript", right_on="contig", suffixes=("_readlength", "_polyA"))
df
```

```python
df = df.merge(hl_df, how="inner", left_on="transcript", right_on="Transcript.stable.ID", suffixes=("", "_hl"))
df
```

```python
df.columns
```

```python
cols = ["transcript", "name", "Gene.stable.ID", "n_readlength", "mean_readlength", "sigma_readlength", "prop_readlength", "min_readlength", 'n_polyA', 'mean_polyA', 'sigma_polyA', 'prop_polyA',
       'min_polyA', "t5"]
df = df[cols]
df
```

```python
df = df.rename(columns = {"Gene.stable.ID" : "gene","mean_readlength" : "maxmean_readlength", "mean_polyA" : "maxmean_polyA"})
```

```python
fpath = save_path
df.to_csv(fpath, index=False)
```

## mmu_dRNA_3T3_mion_1

```python
### experiment mmu_dRNA_3T3_mion_1 ### 
halflives_txt_path = './data/Eisen/Eisen_HL_cleaned.txt'
nanoplen = './experiments/mmu_dRNA_3T3_mion_1/nanoplen_length_synth_rate.tab'
polyA = './experiments/mmu_dRNA_3T3_mion_1/polyA_synth_rate.tab'
gene_transcripts = "./data/mm10/gene_transcript_ids.txt"
save_path = './experiments/mmu_dRNA_3T3_mion_1/features_v1.csv'
```

```python
import pandas as pd

df_readlength = pd.read_csv(nanoplen,sep='\t',header=(0))
print(len(df_readlength))
df_readlength = df_readlength[df_readlength['conv'] != 0]
print(len(df_readlength))
df_readlength
```

```python
df_polyA = pd.read_csv(polyA,sep='\t',header=(0))
print(len(df_polyA))

df_polyA = df_polyA[df_polyA['conv'] != 0]
print(len(df_polyA))

df_polyA
```

```python
cols = ["refname", "genename", "t5"]
hl_df = pd.read_csv(halflives_txt_path, sep='\t', names=cols)#, skiprows=(0,1), header=(2))
hl_df[:10]
```

```python
conv = pd.read_csv(gene_transcripts, sep="\t", header=0)
conv
```

```python
hl_df = hl_df.merge(conv, left_on="genename", right_on="Gene name", how="inner")
hl_df = hl_df.rename(columns={"Gene stable ID":"Gene.stable.ID","Transcript stable ID":"Transcript.stable.ID","Gene name":"name"})
hl_df
```

```python
df = df_readlength.merge(df_polyA, how="inner", left_on="transcript", right_on="contig", suffixes=("_readlength", "_polyA"))
df
```

```python
df = df.merge(hl_df, how="inner", left_on="transcript", right_on="Transcript.stable.ID", suffixes=("", "_hl"))
df
```

```python
df["Gene.stable.ID"].value_counts()
```
